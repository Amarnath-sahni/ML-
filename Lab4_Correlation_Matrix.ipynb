```python
import numpy as np
import pandas as pd

df= pd.read_csv('Data (2).csv')
# 1. Decimal Scaling Normalization
def decimal_scaling(df)
    max_val = np.max(np.abs(df["Pregnant"]))
    scaling_factor = 10 ** np.ceil(np.log10(max_val))
    return df / scaling_factor

# 2. Min-Max Normalization
def min_max_normalization(df):
    min_val = np.min(df)
    max_val = np.max(df)
    return (df - min_val) / (max_val - min_val)

# 3. Z-Score Normalization (Zero-Mean Normalization)
def z_score_normalization(df):
    mean = np.mean(df)
    std_dev = np.std(df)
    return (df - mean) / std_dev

# Applying all normalization methods
decimal_scaled_data = decimal_scaling(df)
min_max_normalized_data = min_max_normalization(df)
z_score_normalized_data = z_score_normalization(df)

# Display results
print("Original Data: ", df)
print("Decimal Scaling Normalized Data: ", decimal_scaled_data)
print("Min-Max Normalized Data: ", min_max_normalized_data)
print("Z-Score Normalized Data: ", z_score_normalized_data)
```
